```python
import tensorflow.keras
from PIL import Image, ImageOps
import numpy as np
import cv2
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Load the model../input/nail-disease-final-data/nail diseases 11/bluish nail/18.PNG
model = tensorflow.keras.models.load_model('../input/nddnew/keras_model.h5')

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1.
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)


# Replace this with the path to your image
image = Image.open('../input/nail-disease-final-data/nail diseases 11/bluish nail/18.PNG').convert('RGB')
#resize the image to a 224x224 with the same strategy as in TM2:
#resizing the image to be at least 224x224 and then cropping from the center

size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)

#turn the image into a numpy array
image_array = np.asarray(image)

# display the resized image
#image.show()

# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1

# Load the image into the array
data[0] = normalized_image_array

# run the inference
result = model.predict(data)
#print(Result)
itemindex = np.where(result==np.max(result))

reverse_mapping=['aloperia areata',"beau's lines",'bluish nail'
                 ,'clubbing',"Darier's disease",'eczema',
                 "half and half nailes (Lindsay's nails)",
                 'koilonychia','leukonychia',"Muehrck-e's lines",
                 'onycholyci','red lunula','splinter hemmorrage',
                 "terry's nail",'white nail','yellow nails',
                 'Healthy Nail']
prediction_name = reverse_mapping[itemindex[1][0]]
#print(prediction_name)


#printing result
itemindex = np.where(result==np.max(result))
Accuracy=str(np.max(result*100))
print("Predicted Symptome is " + prediction_name + " with an accuracy of: " +Accuracy)
if prediction_name == 'Healthy Nail' :
    {
    print("Possible diseases for the symtoms is: NONE")

    }
elif prediction_name == 'bluish nail':
    {
        print("Possible diseases are Asthma, Heart disease, Methhemoglobinemia, Polycythemia, Raynauds disease ")
    }
elif prediction_name == "half and half nailes (Lindsay's nails)":
    {
        print("Possible diseases are Psoriasis Drugs like 5 fluorouracil")
    }
elif prediction_name == 'white nail':
    {
        print("Possible diseases are Anemia, Iron deficiency, Liver cirrhosis, Zinc deficiency ")
    }
elif prediction_name == 'red lunula':
    {
        print("Possible diseases are Systemic lupus erythematosis, Cirrhosis. Alopecia aerate, Psoriasis")
    }
elif prediction_name == 'splinter hemmorrage':
    {
        print("Possible diseases are Endocarditis, Leukemia, Scurvy, Vasuliits, Polyarteririts")
    }

elif prediction_name == "terry's nail":
    {
        print("Possible diseases are Cirrhosis of liver, Congestive cardiac failure ")
    }

elif prediction_name == 'yellow nails':
    {
        print("Possible diseases are Bronchitis, Asthma, Pleural effusion, COPD, Hypothyroidism, Brochiectasis, Dpencillamine therapy ")
    }
elif prediction_name == 'clubbing':
    {
        print("Possible diseases are Lung cancer, Lung abscess, Pulmonary tuberculosis, Congestive cardiac failure, Infective endocarditis")
    }
elif prediction_name == 'koilonychia':
    {
        print("Possible diseases are iron deficiency anemia")
    }
elif prediction_name == "beau's lines":
    {
        print("Possible diseases are zinc deficiency, pneumonia ")
    }
elif prediction_name == "Muehrck-e's lines":
    {
        print("Possible diseases are arsenic poisoning, thallium poisoning, hodgkins lymphoma, sickle cell, anemia cardiac failure")
    }
```
